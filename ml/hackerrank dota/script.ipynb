```python
from sklearn import tree
```

# Gathering the data from txt file

```python
theList = list()
with open("trainingdata.txt","r") as f:
    for line in f:
        theList.append(line.strip())
    
f.close()
# print(len(theList))
```

```python
sampleList = theList[0:5]
```

# Categorical to Numerical data mapping via dict()


```python
db = dict()
id = 0
for item in theList:
    listItem = item.split(',')
    listItem.pop()
    for item1 in listItem:
        if item1 not in db:
            db[item1] = id
            id+=1
            
# Now each Charachter of the game is assigned with a id.
```

# Now that we have id's, we shall make the data set of winners and all the players

```python
herosSelected = list()
dataWinner = list()

for players in theList:
    playerList = list()
    heros = players.split(',')
    dataWinner.append(heros.pop())
    for hero in heros:
        playerList.append(db.get(hero))
    herosSelected.append(playerList)
    
```

```python
dataWinner
```

# Now train the model Decision tree classifier

```python
classifier = tree.DecisionTreeClassifier()
classifier.fit(herosSelected, dataWinner)
```

```python
testList = list()
testOutput = list()
with open("testData.txt",'r') as f:
    for line in f:
        # print(line)
        tempList = list()
        items = line.strip().split(',')
        testOutput.append(items.pop())
        for val in items:
            tempList.append(db.get(val))
        testList.append(tempList)
f.close()

# testOutput
```

```python
predictions = classifier.predict(testList).tolist()
# predictions

count = 0
for i in range(len(predictions)):
    if testOutput[i] == predictions[i]:
        count+=1
print(f'Accuracy of test is {(count/len(predictions))*100}')
```
